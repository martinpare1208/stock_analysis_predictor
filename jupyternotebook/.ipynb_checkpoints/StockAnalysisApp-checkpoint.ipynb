```python
from yahoo_fin.stock_info import *
import datetime as dt
ticker = input('Please select a ticker symbol: ')
end = dt.datetime.now()
start = end - dt.timedelta(days=100)

print(start)
print(end)
```

```python
print(get_live_price(ticker))
```

```python
data = get_data(ticker, start_date = start, end_date = end, index_as_date = True, interval = '1d')
print(data)
```

After getting your intended stock data, we will plot these points onto a graph to better analyze all of this data.

```python
import matplotlib.pyplot as plt
import pandas as pd
plot_point = input(f'Choose which plot point to graph (open, high, low, close, adjclose): ').lower()
```

```python
df = data
df[f'{plot_point}'].plot(label=f'{plot_point.capitalize()} Prices')
plt.xlabel('Date')
plt.ylabel(f'{plot_point.capitalize()} Price')
plt.yscale('linear')
plt.title(f'Stock {plot_point.capitalize()} Prices ({ticker})')
plt.show()
```
